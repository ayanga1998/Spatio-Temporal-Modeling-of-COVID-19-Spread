Dataset Description:

A dataset of mobility and history of reported cases of COVID-19 in England NUTS3 regions, from 3 March to 12 of May. The dataset is segmented in days and the graph is directed and weighted. The graph indicates how many people moved from one region to the other each day, based on Facebook Data For Good disease prevention maps. The node features correspond to the number of COVID-19 cases in the region in the past **window** days. The task is to predict the number of cases in each node after 1 day. For details see this paper: `"Transfer Graph Neural Networks for Pandemic Forecasting." <https://arxiv.org/abs/2009.08388>`

Node Features: The number of reported cases in the given region in the past

Edge Index: Regions in which people traveled each day (Source region -> Target region)

Edge Attributes: The number of people traveling from region to region

y: number of reported cases in each region at the next time step (target variable)

```python
from torch_geometric_temporal.signal import temporal_signal_split
train_dataset, test_dataset = temporal_signal_split(dataset, train_ratio=0.7)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric_temporal.nn.recurrent import A3TGCN

class TemporalGNN(torch.nn.Module):
    def __init__(self, node_features, periods):
        super(TemporalGNN, self).__init__()
        # Attention Temporal Graph Convolutional Cell
        self.tgnn = A3TGCN(in_channels=node_features, 
                           out_channels=32, 
                           periods=periods)
        # Equals single-shot prediction
        self.fc = nn.Sequential(nn.Linear(32, 16),
                               nn.ReLU(),
                               nn.Linear(16, 8),
                               nn.ReLU(),
                               nn.Linear(8,1))

    def forward(self, x, edge_index, edge_weight):
        """
        x = Node features for T time steps
        edge_index = Graph edge indices
        """
        h = self.tgnn(x, edge_index, edge_weight)
        h = self.fc(h)
        return h

TemporalGNN(node_features=1, periods=1)
```

```python
# GPU support
device = torch.device('cpu') # cuda
subset = 2000

# Create model and optimizers
model = TemporalGNN(node_features=1, periods=3).to(device)

num_epochs = 20
history = {"Train Loss": [], "Test Loss": []}
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

print("Running training...")
for epoch in range(num_epochs): 
    train_loss = 0
    test_loss = 0 
    
    model.train()
    for snapshot in train_dataset:
        snapshot = snapshot.to(device)
        x = snapshot.x.reshape(129, 1, 3)
        y = snapshot.y.reshape(129, 1)
        
        # Get model predictions
        y_hat = model(x, snapshot.edge_index, snapshot.edge_attr)
        cur_train_loss = criterion(y_hat, y)
        
        cur_train_loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
        # Mean squared error
        train_loss += cur_train_loss
    
    model.eval()
    with torch.no_grad():
        for snapshot in test_dataset:
            x = snapshot.x.reshape(129, 1, 3)
            y = snapshot.y.reshape(129, 1)
            
            y_hat = model(x, snapshot.edge_index, snapshot.edge_attr)
            cur_test_loss = criterion(y_hat, y)
            
            test_loss += cur_test_loss
    
    train_loss = train_loss/len(list(train_dataset))
    test_loss = test_loss/len(list(test_dataset))
    
    history['Train Loss'].append(train_loss)
    history['Test Loss'].append(test_loss)
    print(f'Epoch: {epoch+1}/10, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}')
```

```python
plt.plot(history['Train Loss'], label = 'Train Loss')
plt.plot(history['Test Loss'], label = 'Test Loss')
plt.legend()
plt.show()
```

```python
model.eval()

# Store for analysis
predictions = []
labels = []
with torch.no_grad():
    for snapshot in test_dataset:
        snapshot = snapshot.to(device)
        x = snapshot.x.reshape(129,1,3)
        # Get predictions
        y_hat = model(x, snapshot.edge_index, snapshot.edge_attr)
        # Mean squared error
        # Store for analysis below
        labels.append(snapshot.y)
        predictions.append(y_hat.cpu())
```
